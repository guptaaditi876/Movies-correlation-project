```python
#Importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
#ggplot is an open source data visualisation package in R and used for better visualisation in python
plt.style.use('ggplot')
#the figure module the figure which contains all the plot elements
from matplotlib.pyplot import figure

#This function makes the plot be appear and be stored inside the jupyter notebook
%matplotlib inline
#Adjusts the configuration of the plots we will create
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#Reading the data 

df = pd.read_csv("movie.csv")
df.head()
```

```python
#Checking missing data's percentage

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} : {} %'.format(col,pct_missing))
```

```python
#Dropping all the missing values

df=df.dropna()
df.head()
```

```python
#Rechecking the datset for missing values

df.isnull().sum()
```

```python
#Checking datatypes

df.dtypes
```

```python
#Changing data types of certain columns

df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
df.head()
```

```python
#Create correct year column. Here pat function is used to extract the year from released column.

df['year_correct']=df['released'].str.extract(pat='([0-9]{4})').astype(int)
df.head()
```

```python
#Here we have sorted the data by the gross column and returned a copy of the original data using inplace= false

df=df.sort_values(by = ['gross'],inplace= False, ascending = False)
df.head()
```

```python
#Changing the default view for the dataset to display max number of rows.Now the above dataset is fully visisble.

pd.set_option('display.max_rows',20)
```

```python
#Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
#We are trying to form correlation of certain parameters like budget and gross

plt.scatter(x=df['budget'], y=df['gross'])
#plt.figure(figsize=(20,12))
plt.title("Gross vs Budget")
plt.ylabel("Gross")
plt.xlabel("Budget")
plt.show()
```

```python
df.head()
```

```python
#Plot the budget vs gross using seaborn

sns.regplot(x='budget',y='gross', data =df, scatter_kws ={"color":"red"}, line_kws = {"color" :"blue"})
```

```python
#lets look at our correlation with the help of df.corr() function

df.corr(method = 'pearson') #types of correlation like pearson,kendall,spearman

#We have found out that there is a high correlation between budget and gross
#Pearson correlation: It is said to be one of the best methods as it gives magnitude as well as direction of correlation
#Kendall correlation: It is a test of strnegth of dependance and not both entities equally correlated.
#Spearman correlation : used to determine the strength of the links between 2 correlations
```

```python
corr_matrix = df.corr(method ='pearson')

sns.heatmap(corr_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

```python
#Now lets look at Company

df.head()

#Now we need to make a correlation between the non numeric values like comapny. For this we will be using cat codes. 
#Cat codes are random values assigned to certain objects, ex: Marvel Studios = 54.
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype("category")
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
df.head()
```

```python
corr_matrix_2 = df_numerized.corr(method = 'pearson')

sns.heatmap(corr_matrix_2, annot = True)

plt.title("Correlation matrix for Numeric Features")

sns.set(rc= {'figure.figsize':(30,12)})

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```
